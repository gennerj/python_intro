# ¿Qué es la indexación en Pandas?
Indexar en Pandas significa seleccionar filas y columnas particulares de un `DataFrame`

Hay dos métodos diferentes de indexación en Pandas:
* loc - selección basada en etiquetas
* iloc - selección basada en índices

# Selección basada en índices
La selección basada en índices consiste en seleccionar los datos en función de su posición numérica en el `dataframe`.

* **iloc** se utiliza para seleccionar los datos en función de la posición numérica
* La sintaxis para utilizar el operador iloc es df.iloc[ ], donde df es un nombre del `dataframe`.
* Puedes pasar las posiciones numéricas de las filas y columnas a seleccionar en el corchete.
* Manos a la obra -> Importa la librería Pandas y carga el archivo 'exam_scores.csv'

```
import pandas as pd

exam_scores = pd.read_csv("exam_scores.csv")

exam_scores.head(4)
```

# Selección basada en etiquetas
* La selección basada en etiquetas selecciona los datos basándose en los nombres/índices de las columnas o filas.
* Esto es importante cuando se seleccionan datos de un `dataframe`
* La selección basada en etiquetas se realiza con loc.
* Loc se utiliza para seleccionar datos basados en el valor/nombre del índice de datos, no en las posiciones numéricas.
* La sintaxis de loc es similar a la de iloc: df.loc[ ], df es un `dataframe`.

image.png

```
exam_scores.loc[0,"race/ethnicity"]
```

```
exam_scores.loc[0:5,["gender","lunch","math score"]]
```

# Ordenando Datos

* Supongamos que somos un profesor y que tenemos las notas de sus alumnos en el `dataframe` 'exam_scores'. Queremos obtener información sobre todos los alumnos cuyo rendimiento es muy malo en "matemáticas".

* ¿Qué tal si revisamos todos los registros del conjunto de datos y averigua manualmente los alumnos cuya puntuación en matemáticas es muy mala?

* Puede pensar en hacer esta tarea manual si tenemos 10 o 20 registros. Pero en el `dataframe` tenemos 1000 registros. Hay que encontrar una manera fácil de hacerlo. Aquí viene la importancia de la técnica de ordenación.

* Los datos en el dataframe 'exam_scores' se orden por default por el índice y no en un orden de alguna columna.

* **Pandas** proporciona un método llamado sort_values() que devuelve el resultado ordenado por los valores de una columna.

* Digamos que queremos obtener la información de los estudiantes que están en orden creciente de puntuación en matemáticas.

```
exam_scores.sort_values(by="math score") # se ordena por default de manera ascendente ( de menor a mayor )
```

```
exam_scores.sort_values(by="math score").head() #podemos combinar funciones y métodos
```

image.png

Podemos obtener el resultado ordenado en orden descendente (de mayor a menor) pasando el parámetro 'ascending' como False en el método sort_values().

```
exam_scores.sort_values(by="math score",ascending=False)
```

# Renombrando índices y columnas

* Renombrar significa cambiar el nombre.

* La mayoría de las veces tenemos conjuntos de datos en los que los nombres de las columnas no son adecuados.

* Por ejemplo, en este `dataframe`, *math score*, *reading score* y *writing score* contienen espacios, por lo que no podemos utilizar el método de selección de atributos (puntos) para seleccionar una columna concreta.

* Pandas proporciona una función 'rename()' para renombrar columnas/índices en un `dataframe`.

* Vamos a renombrar todos los nombres de columnas de nuestro `dataframe` que contengan espacios.

* También cambiaremos el nombre de la columna *'race/ethnicity'* a 'raza'.

image.png

```
exam_scores.rename(columns={'race/ethnicity':'race',
                            'parental level of education':'parent_education_level',
                            'test preparation course':'test_prep_course',
                            'math score':'math_score',
                            'reading score':'reading_score',
                            'writing score':'writing_scor'}, inplace=True)
```

# Datos faltantes ( missing data )

Si en alguna fila o columna de un `dataframe`, el valor no está disponible, se dice que es el valor faltante.

Así se definen los datos perdidos: Los datos faltantes se definen como el valor de los datos que no están almacenados en una columna o fila.

Considere este pequeño conjunto de datos que tiene algunos valores faltantes:

```
df = pd.read_csv("starbucks_location.csv")
```

```
df.head()
```

Pandas proporciona las funciones isnull(), isna() para detectar los valores perdidos. Ambas hacen lo mismo.

df.isna() o df.isnull() devuelve un `dataframe` con valores booleanos que indican los valores perdidos.

```
df.isnull()
```

Podemos obtener el recuento por columnas de todos los valores perdidos utilizando la función de agregación sum()

```
df.isnull().sum()
```

Pandas también proporciona el método fillna() para rellenar los valores perdidos. fillna() proporciona muchas estrategias diferentes para rellenar los valores perdidos.

```
df["City"].fillna("no disponible")
```

```
df["City"].isnull().sum() #porque me siguen apareciendo los valores nulos
```

```
df["City"].fillna("no disponible", inplace=True) #le agrego el parámetro inplace=True
```

```
df["City"].isnull().sum() #ya no tengo valores nulos
```
